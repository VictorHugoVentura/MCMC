```python
from random import random
import matplotlib.pyplot as plt
```

```python
def sample():
    k = 0
    s = 0
    while True:
        s += random()
        k += 1
        if s >= 1:
            break
    return k

def sample_average(n):
    a = 0
    for i in range(1, n + 1):
        s = sample()
        a = (s + (i - 1) * a) / i
    return a
```

```python
n = 1000
init = 1
plot_num = 10

for j in range(plot_num):
    ls = []
    a = 0
    for i in range(1, n + 1):
        s = sample()
        a = (s + (i - 1) * a) / i
        ls.append(a)
    plt.plot(range(init, n + 1), ls[init - 1:])
plt.axhline(y=2.71828, color='r', linestyle='-')
plt.xlabel('Tamanho da amostra')
plt.savefig('aproximacao_de_e.png')
plt.show()
```
